### 스케일링
- 데이터의 스케일(간격)을 맞추는 작업이다.
- 각 컬럼별로 가지는 값의 범위가 다름 -> 이는 같은 1의 차이라도 종속변수에 미치는 영향력이 다르다는 문제 발생
- 이러한 문제를 해결하기 위해 스케일링 시행

- 스케일링 종류
    + 표준화 스케일링: 평균이 0이 되고, 표준편착 1이 되도록 데이터를 고르게 분포시키는데 사용
    + 로버스트 스케일링: 데이터에 아웃라이어(이상치)가 존재하고, 그 영향력을 그대로 유지하고 싶을때 사용
    + 최소-최대 스케일링: 데이터 분포의 특성을 최대한 그대로 유지하고 싶을 때 사용
    + 정규화 스케일링: 행 기준의 스케일링이 필요할 때 사용하지만, 실제로는 거의 사용하지 않음

- 표준화 스케일링: 기존 데이터가 정규분포를 따르며 아웃라이어가 없을 때 무난하게 사용 (평균과 표준편차 사용)
- 로버스트 스케일링: 아웃라이어의 영향을 덜 받고 싶을 때 사용 (사분위수 사용)
- 최대-최소 스케일링: 기존 데이터의 분포를 최대한 유지하고 있을 때 사용 (최대, 최소값 사용)

- 머신러닝에 스케일링 적용시 주의사항
    1. 스케일링 대상에 종속변수를 제외 -> 종속변수가 다른 값으로 바꾸기 때문
    2. 스케일링 전에 훈련셋과 시험셋으로 나누어야 함 -> 시험셋의 데이터는 새로운 데이터이다. 따라서 매번 새롭게 학습하는 프로세스는 비효율적이기 때문


- 스케일링 적용 (최대-최소 스케일러 사용)

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data.drop('class', axis = 1), data['class'], test_size = 0.2, random_state = 100)
```

```python
mm_scaler = MinMaxScaler()
mm_scaler.fit(X_train)
```

```python
X_train_scaled = mm_scaler.transform(X_train)
X_test_scaled = mm_scaler.transform(X_test)
```

- 다른 방법

```python
mm_scaler = MinMaxScaler()
X_train_scaled = mm_scaler.fit_transform(X_train)
X_test_scaled = mm_scaler.fit_transform(X_test)
```

### 모델링 및 예측/평가

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
knn.fit(X_train_scaled, y_train)
pred = knn.predict(X_test_scaled)
```

```python
pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, pred)
```

### 하이퍼파라미터 튜닝

```python
knn = KNeighborsClassifier(n_neighbors = 7)
knn.fit(X_train_scaled, y_train)
pred = knn.predict(X_test_scaled)
accuracy_score(y_test, pred)
```

```python
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train_scaled, y_train)
pred = knn.predict(X_test_scaled)
accuracy_score(y_test, pred)
```

- 일일이 모든 값을 수작업하기엔 번거롭다. for문 사용

```python
for i in range(1, 21):
    knn = KNeighborsClassifier(n_neighbors = i)
    knn.fit(X_train_scaled, y_train)
    pred = knn.predict(X_test_scaled)
    print(accuracy_score(y_test, pred))
```

```python
scores = []
for i in range(1, 21):
    knn = KNeighborsClassifier(n_neighbors = i)
    knn.fit(X_train_scaled, y_train)
    pred = knn.predict(X_test_scaled)
    acc = accuracy_score(y_test, pred)
    scores.append(acc)
```

```python
scores
```

- lineplot으로 시각화

```python
sns.lineplot(x = range(1, 21), y = scores)
```

- 13이후 더 나은 개선은 없다. 따라서 13을 선택하는 것이 가장 합리적이다.

```python
knn = KNeighborsClassifier(n_neighbors = 13)
knn.fit(X_train_scaled, y_train)
pred = knn.predict(X_test_scaled)
accuracy_score(y_test, pred)
```
