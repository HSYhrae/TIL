## 행추가


```python
df = pd.DataFrame({
    "이름": ["Evan", "Sara", "Chris", "Messi", "Woods", "Jordan"], 
    "국어": [70, 90, None, None, 100, 85], 
    "영어": [65, 100, None, 90, 70, 80]
})

df
```

```python
df.loc[6, '이름':'영어'] = ['Kim', 50, 60]
df
```

```python
%%timeit
# 국어점수가 90점 이상이면 'A'
# 국어점수가 80점~89점 'B'
# 국어점수가 70점~79점 'C'
# 나머지는 F
def get_grade(score):
    if score is None:
        return None
    elif score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    else:
        return 'F'
df['등급'] = df['국어'].apply(get_grade)
df
```

```python
%%timeit
df['등급'] = df['국어'].apply(lambda score: 
                             'A' if score is not None and score >= 90
                             else 'B' if score is not None and score >= 80
                             else 'C' if score is not None and score >= 70
                             else 'F')
df
```

- numpy.where, 조건식 한번만 적용
if 조건식, 참일 때, 거짓일 때 

```python
# np.where(조건식, 참, np.where(조건식, 참, (np.where(조건식, 참))))
# np.select(다중조건식, 다중결과값, default-> else 조건)
```

```python
# %%timeit
# 강사님이 한 방식
condition_list = [(df["국어"] >= 90), (df['국어'] >= 80) & (df['국어'] < 90), (df['국어'] >= 70) & (df['국어'] < 80)]
choice_list = ["A", "B", "C"]
df['점수'] = np.select(condition_list, choice_list, default = "F")
df
```

## 코로나 관련 변수만 추출
- 결측값이 있는지 확인
- 결측값이 있는 국가 하나 존재
- totvars 사용

```python
covidtotals.loc[covidtotals[totvars].isnull().sum(axis=1) >= 1, totvars].T
```

```python
covidtotalsonly = covidtotals.loc[:, totvars]
covidtotalsonly['total_cases_q'] = pd.qcut(covidtotalsonly['total_cases'],
                                          labels = ['매우낮음', '낮음', '중간', '높음', '매우높음'],
                                          q=5, precision=0)

covidtotalsonly['total_deaths_q'] = pd.qcut(covidtotalsonly['total_deaths'],
                                          labels = ['매우낮음', '낮음', '중간', '높음', '매우높음'],
                                          q=5, precision=0)
covidtotalsonly.head(1)
```

```python
pd.crosstab(covidtotalsonly['total_cases_q'], covidtotalsonly['total_deaths_q']) 
# 중앙에서부터 2칸이상 떨어진 것은 이상치로 볼 수 있다.
```

```python
covidtotals.loc[(covidtotalsonly.total_cases_q == '매우높음') & (covidtotalsonly.total_deaths_q == '중간'), :].T
```
